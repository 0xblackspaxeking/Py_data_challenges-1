````python
#Challenge #5 - The math to function library
'''
The Concept: Translating formulas into code is the backbone of machine learning.
•	The Task: Create a script that acts as a "Geometry Toolbox."
•	The Logic:
	1.	Define three functions: calc_area_circle(radius), calc_area_square(side), and calc_area_triangle(base, height).
	2.	Use the math module (import math) to get the value of ‣‣ for your circle calculation (‣‣‣‣).
	3.	Create a "Menu" using input() where the user types which shape they want to calculate, and the script calls the correct function.

'''

import math

def main():

  print('Geometry Toolbox')

  print('1. Circle')
  print('2. Square')
  print('3. Triangle')

  choice = input('Enter your choice: ')
  def calc_area_circle(): # Removed 'radius' parameter
    radius = float(input('Enter the radius of the circle: '))
    return math.pi * radius ** 2

  def calc_area_square(): # Removed 'side' parameter
    side = float(input('Enter the side length of the square: '))
    return side ** 2

  def calc_area_triangle(): # Removed 'base, height' parameters
    base = float(input('Enter the base length of the triangle: '))
    height = float(input('Enter the height of the triangle: '))
    return 0.5 * base * height

  if choice == '1':
    area = calc_area_circle()
    print(f'The area of the circle is: {area:.2f}')
  elif choice == '2':
    area = calc_area_square()
    print(f'The area of the square is: {area:.2f}')
  elif choice == '3':
    area = calc_area_triangle()
    print(f'The area of the triangle is: {area:.2f}')
  else:
    print('Invalid choice')

main()
````

### Drone Data Structure (JSON)

Here's a proposed JSON structure for the drone data:

```json
{
  "Timestamp": "YYYY-MM-DDTHH:MM:SS.ffffff",
  "Latitude": "float",
  "Longitude": "float",
  "Battery": "integer (percentage)",
  "Speed": "float (e.g., m/s)",
  "Logs": "string (e.g., 'Normal operation', 'Sensor error')"
}
```

Now, let's create a script that simulates a drone generating this data every 2 seconds and saves it into a SQLite database.

````python
import sqlite3
import datetime
import time
import random

def initialize_db(db_name='drone_data.db'):
    """Initializes the SQLite database and creates the drone_readings table."""
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS drone_readings (
            timestamp TEXT,
            latitude REAL,
            longitude REAL,
            battery INTEGER,
            speed REAL,
            log_message TEXT
        )
    ''')
    conn.commit()
    conn.close()
    print(f"Database '{db_name}' initialized and 'drone_readings' table created (if not exists).")

def generate_drone_data():
    """Generates a dictionary with simulated drone data."""
    timestamp = datetime.datetime.now().isoformat()
    latitude = round(random.uniform(34.0, 34.5), 6)  # Example latitude range
    longitude = round(random.uniform(-118.0, -118.5), 6) # Example longitude range
    battery = random.randint(20, 100) # Battery percentage
    speed = round(random.uniform(0.5, 20.0), 2) # Speed in m/s
    logs = random.choice([
        "Normal operation",
        "Minor turbulence detected",
        "Altitude stabilized",
        "Sensor check complete",
        "Battery level nominal"
    ])
    return {
        'Timestamp': timestamp,
        'Latitude': latitude,
        'Longitude': longitude,
        'Battery': battery,
        'Speed': speed,
        'Logs': logs
    }

def insert_data(data, db_name='drone_data.db'):
    """Inserts a single drone data record into the database."""
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO drone_readings (timestamp, latitude, longitude, battery, speed, log_message)
        VALUES (?, ?, ?, ?, ?, ?)
    ''',
    (data['Timestamp'], data['Latitude'], data['Longitude'], data['Battery'], data['Speed'], data['Logs']))
    conn.commit()
    conn.close()

# Initialize the database
initialize_db()

print("Simulating drone data generation and saving to SQLite (running for 10 cycles)...")

# Simulate data generation and saving
num_cycles = 10 # Let's run for 10 cycles for demonstration
for i in range(num_cycles):
    drone_reading = generate_drone_data()
    insert_data(drone_reading)
    print(f"[{i+1}/{num_cycles}] Data logged: {drone_reading['Timestamp']}")
    time.sleep(2) # Simulate data generation every 2 seconds

print("Simulation complete. Data saved to drone_data.db")
````

Let's display the first few records from the `drone_data.db` to verify the data was saved correctly.

````python
import sqlite3
import pandas as pd

def fetch_data(db_name='drone_data.db', limit=5):
    """Fetches and displays data from the drone_readings table."""
    conn = sqlite3.connect(db_name)
    query = f"SELECT * FROM drone_readings ORDER BY timestamp DESC LIMIT {limit}"
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df

display(fetch_data())
````
